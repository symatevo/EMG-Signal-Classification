```python
# Redefining the aesthetics of the plot to highlight models with 90+ accuracy, RF accuracy, and RNN inference time
import matplotlib.pyplot as plt

# Data for accuracies and inference times
models = ['DT', 'RF', 'SVM_rbf', 'SVM_lin', 'SVM_poly', 'RNN']
accuracies = [87.80, 97.56, 82.31, 92.68, 72.25, 91.05]
inference_times = [0.000736, 0.033834, 0.001896, 0.001056, 0.001542, 0.140401]

fig, ax1 = plt.subplots(figsize=(12, 7))

# Setting colors for the bars and highlights
default_bar_color = '#1f77b4'
highlight_accuracy_color = '#2ca02c'  # For models with 90+ accuracy
rf_highlight_color = 'gold'  # For RF model
rnn_highlight_color = 'magenta'  # For RNN inference time

# Plotting accuracies
ax1.set_xlabel('Models', fontsize=14)
ax1.set_ylabel('Accuracy (%)', color='tab:blue', fontsize=14)
bars = ax1.bar(models, accuracies, color=default_bar_color)
ax1.tick_params(axis='y', labelcolor='tab:blue', labelsize=12)
ax1.tick_params(axis='x', labelsize=12)

# Highlighting models with 90+ accuracy and specifically RF
for i, acc in enumerate(accuracies):
    if acc >= 90:
        bars[i].set_color(highlight_accuracy_color)
    if models[i] == 'RF':
        bars[i].set_edgecolor(rf_highlight_color)
        bars[i].set_linewidth(2)

# Plotting inference times
ax2 = ax1.twinx()
ax2.set_ylabel('Inference Time (s)', color='tab:red', fontsize=14)
inference_line, = ax2.plot(models, inference_times, color='tab:red', marker='o', markersize=8, linewidth=2, linestyle='--')
ax2.tick_params(axis='y', labelcolor='tab:red', labelsize=12)

# Highlighting RNN inference time
ax2.plot(models[-1], inference_times[-1], color=rnn_highlight_color, marker='o', markersize=12)

# Adding a title and grid
plt.title('Model Comparison: Accuracy and Inference Time', fontsize=16)
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)

# Showing the plot
plt.tight_layout()
plt.show()
```
